```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM, GRU
```

```python
## Load the imdb dataset
num_words=10000 # Number of words to keep
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=num_words)

# pad sequences to have the same length
maxlen=100 # Maximum sequence length
x_train = pad_sequences(x_train, maxlen=maxlen, padding="post", truncating="post")
x_test = pad_sequences(x_test, maxlen=maxlen, padding="post", truncating="post")

print(x_train.shape)
x_train[0]
# maxi= [max(x_train[i]) for i in range(101)]
# print(maxi)

# for i in range(101):
#     print(max(x_train[i]))
```

```python
# Build the Bidirectional RNN model

from keras.layers import Input

num_words=10000
embedding_dim=32 # Embedding dimension layer
maxlen=100
model = Sequential([
    Input(shape=(maxlen,)),  # This tells the model to expect sequences of 100 integers
    Embedding(num_words, embedding_dim), # You can remove input_length here now
    SimpleRNN(5, return_sequences=True), # 5 RNN units
    SimpleRNN(5),
    Dense(1, activation='sigmoid') # Binary Classification (positive/negative)
])

model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# train the model
history= model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# Define the Deep LSTM model

from keras.layers import Input

model = Sequential([
    Input(shape=(100,)),  # This tells the model to expect sequences of 100 integers
    Embedding(10000, 32), # You can remove input_length here now
    LSTM(5, return_sequences=True),
    LSTM(5),
    Dense(1, activation='sigmoid')
])

# Display the model architecture 
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=["accuracy"])
```

```python
# train the model
history= model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
```

```python
# Define the Deep GRU model

from keras.layers import Input

model = Sequential([
    Input(shape=(100,)),  # This tells the model to expect sequences of 100 integers
    Embedding(10000, 32), # You can remove input_length here now
    GRU(5, return_sequences=True),
    GRU(5),
    Dense(1, activation='sigmoid')
])

# Display the model architecture 
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss="binary_crossentropy", metrics=["accuracy"])
```
